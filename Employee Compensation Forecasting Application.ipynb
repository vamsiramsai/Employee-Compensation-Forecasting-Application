```python
%%sql
SELECT
  Name,
  Role,
  Location,
  Years_of_Experience AS Experience,
  Compensation,
  ROUND(Compensation * 1.10, 2) AS Updated_Compensation,
  Active
FROM EmployeeData
WHERE Active = 'Y';
```
